```python
# Define a function to read the files and count the number of rows with the same predicted values
def count_same_predicted_values(file1, file2, file3):
    # Open the files
    with open(file1, 'r') as f1, open(file2, 'r') as f2, open(file3, 'r') as f3:
        # Initialize count
        count = 0
        # Iterate over the rows of all files simultaneously
        for line1, line2, line3 in zip(f1, f2, f3):
            # Split the lines to extract the predicted values
            predicted_value1 = line1.strip().split(',')[1]  # Assuming the predicted value is in the second column
            predicted_value2 = line2.strip().split(',')[1]  # Assuming the predicted value is in the second column
            predicted_value3 = line3.strip().split(',')[1]  # Assuming the predicted value is in the second column
            # Compare the predicted values and increment count if they are equal across all files
            if predicted_value1 == predicted_value2 == predicted_value3:
                count += 1
    return count

# File paths
file1 = 'DATA/Submission20.csv'  # Replace 'file1.csv' with the path to your first file
file2 = 'DATA/Submission7.csv'  # Replace 'file2.csv' with the path to your second file
file3 = 'DATA/Submission9.csv'  # Replace 'file2.csv' with the path to your second file

# Count the number of rows with the same predicted values across all files
num_same_predicted_values = count_same_predicted_values(file1, file2, file3)
print("Number of rows with the same predicted values across all files:", num_same_predicted_values)
```

```python
def create_voting_file(file1, file2, file3, output_file):
    # Open the files
    with open(file1, 'r') as f1, open(file2, 'r') as f2, open(file3, 'r') as f3:
        # Open the output file for writing
        with open(output_file, 'w') as out_file:
            # Iterate over the rows of all files simultaneously
            for line1, line2, line3 in zip(f1, f2, f3):
                # Split the lines to extract the predicted values
                predicted_value1 = line1.strip().split(',')[1]  # Assuming the predicted value is in the second column
                predicted_value2 = line2.strip().split(',')[1]  # Assuming the predicted value is in the second column
                predicted_value3 = line3.strip().split(',')[1]  # Assuming the predicted value is in the second column
                # Check if two predicted values are the same
                if predicted_value1 == predicted_value2:
                    # If they are the same, write the predicted value to the output file
                    out_file.write(f"{line1.strip()}\n")
                elif predicted_value1 == predicted_value3:
                    # If predicted_value1 matches predicted_value3, write predicted_value1 to the output file
                    out_file.write(f"{line1.strip()}\n")
                elif predicted_value2 == predicted_value3:
                    # If predicted_value2 matches predicted_value3, write predicted_value2 to the output file
                    out_file.write(f"{line2.strip()}\n")
                else:
                    # If there's no consensus, write 0 as the predicted value to the output file
                    out_file.write(f"{line2.strip()}\n")
 # Assuming the first column is the ID column
```

```python
output_file = 'voting_result3.csv'  # Output file to write the voting-based predicted values

# Create the new file with voting-based predicted values
create_voting_file(file1, file2, file3, output_file)
```
